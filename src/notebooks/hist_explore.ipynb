```python
import cv2
import numpy as np
```

```python
%matplotlib inline
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('test_images/apple_01.JPG')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
```

```python
color = ('r','g','b')
for i,col in enumerate(color):
    histr = cv2.calcHist([img],[i],None,[255],[0,255]) / 2500
    print(histr.shape)
    plt.plot(histr,color = col)
    plt.xlim([0,256])
plt.show()
```

```python
h_red = cv2.calcHist([img],[0],None,[255],[0,255])
h_green = cv2.calcHist([img],[1],None,[255],[0,255])
h_blue = cv2.calcHist([img],[2],None,[255],[0,255])
```

```python
hist = np.concatenate((h_red, h_green, h_blue), 0)
hist /= np.max(hist)
print(np.max(hist))
print(hist.shape)
```
